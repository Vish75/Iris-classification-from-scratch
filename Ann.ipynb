```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from keras import models
from keras import layers
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('train.csv').dropna()
```

```python
df.shape
```

```python
df.head()
```

```python
df['category'].value_counts()/df['category'].shape
```

```python
X = df.drop('category',axis=1)
y = df['category']
```

```python
X
```

```python
y
```

```python
encoder = OneHotEncoder()
```

```python
encoded_Y = encoder.fit(y.values.reshape(-1,1))
encoded_Y = encoded_Y.transform(y.values.reshape(-1,1)).toarray()
encoded_Y
```

```python
encoded_Y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,encoded_Y,test_size = 0.2,random_state =22)
```

```python
sc = StandardScaler()
X_train_new = sc.fit_transform(X_train)
X_test_new = sc.transform(X_test)
```

```python
pca = PCA(n_components= 100)
X_train_trans = pca.fit_transform(X_train_new)
X_test_trans = pca.transform(X_test_new)
X_train_trans.shape
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
input_shape = X_train_trans.shape[1]
n_batch_size = 20
n_steps_per_epoch = int(X_train_trans.shape[0] / n_batch_size)
n_test_steps = int(X_test_trans.shape[0] / n_batch_size)
n_epochs = 25
num_classes = y_train.shape[1]
```

```python
num_classes
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(12, activation='relu', input_shape=(input_shape,)))
```

```python
model.add(layers.Dense(12, activation='relu'))
```

```python
model.add(layers.Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train_trans, y_train, steps_per_epoch=n_steps_per_epoch, epochs=n_epochs, batch_size=n_batch_size)
```

```python
y_pred = model.predict(X_test_trans)
```

```python
y_test
```

```python
re_transformed_y_pred = encoder.inverse_transform(y_pred)
```

```python
re_transformed_y_pred
```

```python
re_transformed_y_test = encoder.inverse_transform(y_test)
```

```python
re_transformed_y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(re_transformed_y_test,re_transformed_y_pred)
```
