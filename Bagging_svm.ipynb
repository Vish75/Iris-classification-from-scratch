```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('train.csv')

X = df.iloc[:,1:4097]
y = df.iloc[:,4097:4098]
X.shape


from sklearn.cluster import KMeans
km = KMeans(n_clusters = 20, init = 'random', n_init = 10, max_iter = 300, tol = 1e-04, random_state = 0)
y_km = km.fit_predict(X)

y_km.shape
X.shape
X = np.c_[X,y_km]

from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
y = lr.fit_transform(y)
y.shape
y

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 22,stratify=y)

sc = StandardScaler()
X_train_new = sc.fit_transform(X_train)
X_test_new = sc.transform(X_test)

from sklearn.model_selection import GridSearchCV
n_params = {
    'n_estimators': [10, 100, 200, 400, 500],
    'max_samples': [0.01, 0.1, 0.2, 0.5, 0.8, 1],
    'max_features': [0.01, 0.1, 0.2, 0.5, 0.8, 1]
}

from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
clf = BaggingClassifier(base_estimator = SVC(kernel='linear'),random_state= 42)
model = GridSearchCV(estimator= clf, param_grid= n_params, verbose = 3)

model.fit(X_train_new,y_train)
```
