```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Heart.csv')
```

```python
df = df.dropna()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['ChestPain'] = label_encoder.fit_transform(df['ChestPain'])
df['Thal'] = label_encoder.fit_transform(df['Thal'])
df['AHD'] = label_encoder.fit_transform(df['AHD'])
```

```python
df
```

```python
new_df = df.iloc[:,1:14]
new_df
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scalar.fit(new_df)
scaled_data = scalar.transform(new_df)
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=9)
pca.fit(scaled_data)
pcaed_data = pca.transform(scaled_data)
```

```python
pcaed_data
```

```python
pcaed_data.shape
```

```python
pca.explained_variance_ratio_
```

```python
exp_var_pca = pca.explained_variance_ratio_
```

```python
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
```

```python
cum_sum_eigenvalues
```

```python
y = df['AHD']
```

```python
y
```

```python
y.shape
```

```python
Processing_data = np.c_[pcaed_data,y]
```

```python
Processing_data
```

```python
new_df = pd.DataFrame(Processing_data, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','Target'])
```

```python
new_df
```

```python
new_df_no = new_df[new_df['Target']==0.0]
```

```python
new_df_no
```

```python
new_df_yes = new_df[new_df['Target']==1.0]
```

```python
new_df_yes
```

```python
new_df_1 = new_df_no.iloc[:,:9]
new_df_1
```

```python
mean_values_no = new_df_1.mean()
mean_values_no
```

```python
new_df_2 = new_df_yes.iloc[:,:9]
```

```python
new_df_2
```

```python
mean_values_yes = new_df_2.mean()
```

```python
cov_new_df1 = new_df_1.cov()
```

```python
cov_new_df2 = new_df_2.cov()
cov_new_df2
```

```python
Sw = 160*cov_new_df1 + 137*cov_new_df2
Sw
```

```python
# Sw inverse calc
Sw = np.linalg.inv(Sw)
Sw.shape
```

```python
mean_diff = mean_values_yes - mean_values_no
```

```python
mean_diff
```

```python
mean_diff.shape
```

```python
W = np.dot(Sw,mean_diff)
```

```python
W.shape
```

```python
pcaed_data
```

```python
pcaed_data.shape
```

```python
projections = np.dot(pcaed_data,W)
```

```python
projections
```

```python
X = projections
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 22)
```

```python
weights = np.zeros((2,))
```

```python
ones = np.ones((237,))
```

```python
X_new = np.c_[ones,X_train]
```

```python
X_new.shape
```

```python
iter = 800
```

```python
alpha = 0.001
```

```python
from numpy import exp
for i in range(iter):
    z = np.dot(X_new,weights)
    p = 1/(1+exp(-z))
    weights = weights - alpha*np.dot(X_new.T,(p-y_train))
```

```python
weights
```

```python
ones_1 = np.ones((60,))
```

```python
X_pred = np.c_[ones_1,X_test]
```

```python
X_pred
```

```python
X_pred.shape
```

```python
y_pred = np.dot(X_pred,weights)
```

```python
y_pred
```

```python
y_pred1 = []
for i in y_pred:
    if i < 0.5:
        y_pred1.append(0)
    else:
        y_pred1.append(1)
```

```python
y_test = y_test.to_numpy()
```

```python
tp, tn = 0,0
for i in range(len(y_pred1)):
    if y_test[i] == 1 and y_pred1[i] == 1:
        tp += 1
    elif y_test[i] == 0 and y_pred1[i] == 0:
        tn += 1
accuracy = (tp+tn)/len(y_pred1)
```
