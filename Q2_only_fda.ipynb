```python
df = pd.read_csv('Heart.csv')
```

```python
df['AHD'].unique()
```

```python
df = df.iloc[:,1:15]
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['ChestPain'] = label_encoder.fit_transform(df['ChestPain'])
df['Thal'] = label_encoder.fit_transform(df['Thal'])
df
```

```python
mean_values_yes = df.groupby('AHD').mean().loc['Yes']
```

```python
mean_values_yes
```

```python
mean_values_no = df.groupby('AHD').mean().loc['No']
```

```python
mean_values_no
```

```python
new_df = df.iloc[:,:13]
```

```python
new_df_1 = df[df['AHD']=='Yes']
new_df_1
```

```python
new_df_1 = new_df_1.iloc[:,:13]
new_df_1
```

```python
new_df_2 = df[df['AHD']=='No']
new_df_2 = new_df_2.iloc[:,:13]
new_df_2
```

```python
cov_new_df1 = new_df_1.cov()
```

```python
cov_new_df1
```

```python
cov_new_df2 = new_df_2.cov()
cov_new_df2
```

```python
Sw = 137*cov_new_df1 + 160*cov_new_df2
Sw
```

```python
# Sw inverse calc
Sw = np.linalg.inv(Sw)
Sw.shape
```

```python
mean_diff = mean_values_yes - mean_values_no
```

```python
mean_diff.shape
```

```python
W = np.dot(Sw,mean_diff)
```

```python
W
```

```python
W.shape
```

```python
new_df
```

```python
projections = np.dot(new_df,W)
```

```python
projections
```

```python
projections.shape
```

```python
X = projections
y = df['AHD'].to_numpy()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 22)
```

```python
X_train.shape
```

```python
weights = np.zeros((2,))
```

```python
weights.shape
```

```python
ones = np.ones((237,))
```

```python
ones
```

```python
X_new = np.c_[ones,X_train]
```

```python
X_new
```

```python
X_new.shape
```

```python
iter = 800
```

```python
alpha = 0.001
```

```python
lr = preprocessing.LabelEncoder()
y_train = lr.fit_transform(y_train)
```

```python
y_train
```

```python
from numpy import exp
for i in range(iter):
    z = np.dot(X_new,weights)
    p = 1/(1+exp(-z))
    weights = weights - alpha*np.dot(X_new.T,(p-y_train))
```

```python
weights
```

```python
X_test.shape
```

```python
ones_1 = np.ones((60,))
```

```python
ones_1.shape
```

```python
X_pred = np.c_[ones_1,X_test]
```

```python
X_pred.shape
```

```python
y_pred = np.dot(X_pred,weights)
```

```python
y_pred.shape
```

```python
y_pred1 = []
for i in y_pred:
    if i < 0.5:
        y_pred1.append(0)
    else:
        y_pred1.append(1)
```

```python
le = preprocessing.LabelEncoder()
y_test = le.fit_transform(y_test)
y_test
```

```python
y_pred1
```

```python
tp, tn = 0,0
for i in range(len(y_pred1)):
    if y_test[i] == 1 and y_pred1[i] == 1:
        tp += 1
    elif y_test[i] == 0 and y_pred1[i] == 0:
        tn += 1
accuracy = (tp+tn)/len(y_pred1)
```
