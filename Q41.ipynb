```python
Sw1 = np.linalg.inv(Sw)
```

```python
Sw1
```

```python
Decom_mat = np.dot(Sw1,Sb)
```

```python
Decom_mat
```

```python
eigen_values, eigen_vectors = np.linalg.eig(Decom_mat)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
eig_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]
```

```python
eig_pairs
```

```python
eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)
```

```python
eig_pairs
```

```python
for i in eig_pairs:
    print(i[0])
```

```python
eigv_sum = sum(eigen_values)
for i,j in enumerate(eig_pairs):
    print('eigenvalue {0:}: {1:.2%}'.format(i+1, (j[0]/eigv_sum).real))
```
