```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.category.value_counts()
```

```python
X = df.iloc[:,1:4097]
y = df.iloc[:,4097:4098]
X.shape
```

```python
df['category'].unique()
```

```python
y
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
pca = PCA(n_components= 100)
X = pca.fit(X).transform(X)
lda = LinearDiscriminantAnalysis(n_components = 19)
X = lda.fit(X,y).transform(X)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 20, init = 'random', n_init = 10, max_iter = 300, tol = 1e-04, random_state = 0)
y_km = km.fit_predict(X)
```

```python
y_km.shape
X.shape
X = np.c_[X,y_km]
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
y = lr.fit_transform(y)
y.shape
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 22,stratify=y)
```

```python
y_df_train = pd.DataFrame(y_train)
y_df_train.value_counts()
y_df_test = pd.DataFrame(y_test)
y_df_test.value_counts()
```

```python
sc = StandardScaler()
X_train_new = sc.fit_transform(X_train)
X_test_new = sc.transform(X_test)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_new,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred1 = clf.predict(X_train_new)
accuracy_score(y_train,y_pred1)
```

```python
y_pred = clf.predict(X_test_new)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
df_new = pd.read_csv('test.csv')
```

```python
df_new.head()
```

```python
X_pred = df_new.iloc[:,1:4098]
```

```python
X_pred
```

```python
X_pred_trans = pca.transform(X_pred)
X_pred_v = lda.transform(X_pred_trans)
yo = km.predict(X_pred_v)
X_pred_vo = np.c_[X_pred_v,yo]
X_pred_no = sc.transform(X_pred_vo)
```

```python
y_pred_trans = clf.predict(X_pred_no)
```

```python
y_pred_trans
```

```python
y_only = lr.inverse_transform(y_pred_trans)
```

```python
new_df = pd.DataFrame(y_only)
new_df.to_csv("submission.csv")
```
