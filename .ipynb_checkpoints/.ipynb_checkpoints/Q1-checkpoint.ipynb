```python
import pandas as pd
from PIL import Image
```

```python

data_folder = os.getcwd()
data_subfolders = [f.path for f in os.scandir(data_folder) if f.is_dir()]
print(data_subfolders)

        
count = 0
pixel_dictionary = {}
labels = []

for i in range(28):
    for j in range(28):
        pixel_dictionary['{}x{}'.format(i,j)] = []
        
pixel_dictionary
```

```python
for data_number_folder in data_subfolders:
    number_file_list = [f.path for f in os.scandir(data_number_folder)]
    
    for number_file in number_file_list:
        im = Image.open(number_file) #Opens 28x28 image file
        pixels = im.load()
        count += 1
        
        if not count %1000:
            print(count)
        
        for x in range(im.size[0]):
            for y in range(im.size[1]):
                pixel_dictionary['{}x{}'.format(x,y)].append(pixels[y,x])
        
        labels.append(number_file[len(data_number_folder)-1])
                
```
