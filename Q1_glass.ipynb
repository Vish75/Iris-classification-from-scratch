```python
# Using mahalanobis distance for outlier calculation:
def mahalanobis(x=None,data = None, cov = None):
    x_mu = x - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left_term = np.dot(x_mu,inv_covmat)
    maha = np.dot(left_term,x_mu.T)
    return maha.diagonal()
df_x = X
df_x['mahalanobis_dis'] = mahalanobis(x=df_x,data=df_x)
df_x.head()
```

```python
arr = X.to_numpy()
```

```python
from scipy.spatial.distance import cityblock
```

```python
dis = []
for i in range(len(arr)):
    dist = []
    for j in range(len(arr)):
        cal_val = cityblock(arr[i],arr[j])
        dist.append([j,cal_val])
    dis.append(dist)
dis
```

```python
len(dis[0])
```

```python
for i in dis:
    i.sort(key=lambda x:x[1])
dis
```

```python
k = 15
```

```python
k_dis = []
for i in dis:
    k_dis.append(i[:k])
```

```python
k_dis
```

```python
k_neighs = []
for i in k_dis:
    k_neigh_i = []
    for j in i:
        k_neigh_i.append(j[0])
    k_neighs.append(k_neigh_i)
k_neighs
```

```python
k_neighs_dis = []
for i in k_dis:
    k_neigh_i_dis = []
    for j in i:
        k_neigh_i_dis.append(j[1])
    k_neighs_dis.append(k_neigh_i_dis)
k_neighs_dis
```

```python
reachibility_dis = []
for i in range(len(k_dis)):
    calc_dis = []
    for j in range(15):
        calc_dis.append(max(k_neighs_dis[i][j],k_neighs_dis[j][14]))
    reachibility_dis.append(calc_dis)
reachibility_dis
```

```python
lrd = []
for i in reachibility_dis:
    x = 0
    for j in i:
        x= x + j
    lrd.append(x/15)
lrd
```

```python
LRD = []
for i in lrd:
    LRD.append(1/i)
```

```python
LRD
```

```python
lof = []
for i in range(len(LRD)):
    x_tot = 0
    for j in range(15):
        x_tot = x_tot + (LRD[k_neighs[i][j]])
    lof.append(x_tot/LRD[i])
lof
```

```python
LOF = []
for i in lof:
    LOF.append(i/15)
LOF
```
